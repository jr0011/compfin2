# 1. Option Pricing: Realized Variance and Carr-Madan

## 1.1 Itô’s Lemma and Log-Asset DynamicsAsset Price Dynamics
### Theory
We assume the underlying asset price $S_t$ follows a continuous diffusion process under the risk-neutral measure $\mathbb{Q}$. Even in the presence of an adapted stochastic volatility process $\sigma_t$, the dynamics are characterized by the following Stochastic Differential Equation (SDE):$$\frac{dS_t}{S_t} = r dt + \sigma_t dB_t$$where $r$ is the risk-free rate and $B_t$ is a standard Brownian motion.
A fundamental property of this process is its quadratic variation, given by $(dS_t)^2 = \sigma_t^2 S_t^2 dt$, which captures the realized variance of the asset.The General Form of Itô’s Lemma To derive the price of a variance claim, we require a transformation of the asset price. For a general function $f(t, S_t)$ that depends on both time and the stochastic price process, Itô’s Lemma provides the differential:$$df(t, S_t) = \frac{\partial f}{\partial t} dt + \frac{\partial f}{\partial S} dS_t + \frac{1}{2} \frac{\partial^2 f}{\partial S^2} (dS_t)^2$$In the context of the log-contract, we set $f(S_t) = \ln(S_t)$. Because the function $\ln(S_t)$ has no explicit dependency on time $t$ (i.e., the payoff structure of the logarithm does not decay or change purely by the passage of time), the partial derivative $\frac{\partial f}{\partial t}$ is zero. This simplifies the expansion by removing the "drift" or "theta" component related to time.

### Proof
Our goal is to demonstrate that the integrated (realized) variance $\int_{0}^{T} \sigma_{u}^{2} du$ can be replicated using a position in the underlying asset and a log-contract:$$\int_{0}^{T} \sigma_{u}^{2} du = 2 \int_{0}^{T} \frac{dS_{u}}{S_{u}} - 2 \ln\left(\frac{S_{T}}{S_{0}}\right)$$
#### Step 1: Differentiation 
Applying the simplified version of Itô’s Lemma (where $\frac{\partial f}{\partial t} = 0$) to our chosen function $f(S_t) = \ln(S_t)$. First, we compute the first and second partial derivatives of $f$ with respect to the asset price $S_t$:$$\frac{\partial f}{\partial S} = \frac{1}{S_t}, \quad \frac{\partial^2 f}{\partial S^2} = -\frac{1}{S_t^2}$$Substituting these derivatives back into the Itô differential formula yields:$$d \ln(S_t) = \frac{1}{S_t} dS_t + \frac{1}{2} \left( -\frac{1}{S_t^2} \right) (dS_t)^2$$
#### Step 2: Substitution of Quadratic Variation 
By substituting the asset’s quadratic variation $(dS_t)^2 = \sigma_t^2 S_t^2 dt$ into the equation above:$$d \ln(S_t) = \frac{dS_t}{S_t} - \frac{1}{2S_t^2} (\sigma_t^2 S_t^2 dt)$$The $S_t^2$ terms cancel out, yielding the fundamental relationship between arithmetic and logarithmic returns:$$d \ln(S_t) = \frac{dS_t}{S_t} - \frac{1}{2} \sigma_t^2 dt$$
#### Step 3: Integration and Rearrangement 
We now integrate both sides over the time horizon $[0, T]$:$$\ln(S_T) - \ln(S_0) = \int_{0}^{T} \frac{dS_u}{S_u} - \frac{1}{2} \int_{0}^{T} \sigma_u^2 du$$By multiplying the entire equation by $-2$ and isolating the integral for realized variance, we obtain:$$\int_{0}^{T} \sigma_u^2 du = 2 \int_{0}^{T} \frac{dS_u}{S_u} - 2 \ln\left(\frac{S_T}{S_0}\right)$$

## 1.2 Valuation and Decomposition of the Variance Claim

### Theory
Valuation and the Log-ContractThe Variance Claim Price ($\Pi_t$) We define the time-$t$ price of the realized variance over the interval $[0, T]$ as the discounted expected value under the risk-neutral measure $\mathbb{Q}$:$$\Pi_t := e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} \left[ \int_{0}^{T} \sigma_u^2 du \;\middle|\; \mathcal{F}_t \right]$$This represents the fair value at time $t$ of a contract that pays the total realized variance at maturity $T$. 

We also define the time-$t$ price of a log-contract. A log-contract is a derivative whose payoff at maturity is $-2 \ln(S_T/S_0)$. Its risk-neutral price at time $t$ is:$$L_t := -2 e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} \left[ \ln\left(\frac{S_T}{S_0}\right) \;\middle|\; \mathcal{F}_t \right]$$
### The Goal 
We want to show that the variance claim $\Pi_t$ can be priced dynamically and admits the following decomposition:$$\Pi_t = L_t + 2(T-t)r e^{-r(T-t)} + 2 e^{-r(T-t)} \int_{0}^{t} \frac{dS_u}{S_u}$$
### Proof: Decomposition of the Variance Claim
#### Step 1: Substitution of the Variance Identity 
We start with the definition of $\Pi_t$. Using the fundamental identity derived in Section 1.1, we can substitute the entire integral of the realized variance $\int_{0}^{T} \sigma_u^2 du$ with its equivalent expression in terms of arithmetic and logarithmic returns:$$\Pi_t = e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} \left[ 2 \int_{0}^{T} \frac{dS_u}{S_u} - 2 \ln\left(\frac{S_T}{S_0}\right) \;\middle|\; \mathcal{F}_t \right]$$
#### Step 2: Linearity of Expectation and Identifying $L_t$ 
Using the linearity property of the expectation operator, we can split the equation into two separate expected values:$$\Pi_t = 2 e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} \left[ \int_{0}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right] - 2 e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} \left[ \ln\left(\frac{S_T}{S_0}\right) \;\middle|\; \mathcal{F}_t \right]$$Notice that the entire second term matches exactly our theoretical definition of the log-contract price. We can substitute it with $L_t$:$$\Pi_t = 2 e^{-r(T-t)} \mathbb{E}^{\mathbb{Q}} \left[ \int_{0}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right] + L_t$$

#### Step 3: Evaluating the Expected Arithmetic 
Returns Now, let's isolate and evaluate the remaining expected value term: $\mathbb{E}^{\mathbb{Q}} \left[ \int_{0}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right]$.We split this integral into two parts: the realized past (from $0$ to $t$, which is known at time $t$ and therefore can be taken out of the expectation) and the unknown future (from $t$ to $T$):$$\mathbb{E}^{\mathbb{Q}} \left[ \int_{0}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right] = \int_{0}^{t} \frac{dS_u}{S_u} + \mathbb{E}^{\mathbb{Q}} \left[ \int_{t}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right]$$Recall the risk-neutral dynamics of the asset: $\frac{dS_u}{S_u} = r du + \sigma_u dB_u$. We substitute this into the expectation of the future returns:$$\mathbb{E}^{\mathbb{Q}} \left[ \int_{t}^{T} (r du + \sigma_u dB_u) \;\middle|\; \mathcal{F}_t \right] = \int_{t}^{T} r du + \mathbb{E}^{\mathbb{Q}} \left[ \int_{t}^{T} \sigma_u dB_u \;\middle|\; \mathcal{F}_t \right]$$Since the Itô integral with respect to a Brownian motion is a martingale, its conditional expectation is zero. The deterministic drift $r$ simply integrates to $r(T-t)$:$$\mathbb{E}^{\mathbb{Q}} \left[ \int_{t}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right] = r(T-t) + 0$$So, the isolated expected value becomes:$$\mathbb{E}^{\mathbb{Q}} \left[ \int_{0}^{T} \frac{dS_u}{S_u} \;\middle|\; \mathcal{F}_t \right] = \int_{0}^{t} \frac{dS_u}{S_u} + r(T-t)$$
#### Step 4: Final Assembly 
Finally, we substitute this evaluated expectation back into the main equation for $\Pi_t$ from Step 2:$$\Pi_t = 2 e^{-r(T-t)} \left( \int_{0}^{t} \frac{dS_u}{S_u} + r(T-t) \right) + L_t$$By rearranging the terms, we achieve the requested decomposition:$$\Pi_t = L_t + 2(T-t)r e^{-r(T-t)} + 2 e^{-r(T-t)} \int_{0}^{t} \frac{dS_u}{S_u}$$

## 1.3 Portfolio Construction
### Theory
To dynamically replicate the realized variance claim $\Pi_t$, we construct a portfolio $V_t$ at time $t \in [0, T]$ consisting of three distinct financial instruments
- The Log-Contract: $1$ unit of the log-contract, with current value $L_t$.
- The Underlying Asset: $\phi_t$ shares of the underlying stock $S_t$.
- The Money-Market Account: $\psi_t$ units of the risk-free bank account $A_t$, which evolves deterministically as $A_t = e^{rt}$.
The total value of this portfolio at time $t$ is the sum of its components:$$V_t = L_t + \phi_t S_t + \psi_t A_t$$
#### The Weights 
We are given the following specific position sizes for the stock ($\phi_t$) and the bank account ($\psi_t$):$$\phi_t := \frac{2e^{-r(T-t)}}{S_t}$$ $$\psi_t := 2e^{-rT} \left( \int_{0}^{t} \frac{dS_u}{S_u} + (T-t)r - 1 \right)$$
#### The Goal 
We want to prove algebraically that the total value of this portfolio, $V_t$, exactly equals the theoretical price of the variance claim $\Pi_t$ that we derived in Section 1.3:$$\Pi_t = L_t + 2(T-t)r e^{-r(T-t)} + 2 e^{-r(T-t)} \int_{0}^{t} \frac{dS_u}{S_u}$$
### Proof: Replication of the Variance Claim
#### Step 1: Valuation of the Stock Position 
We calculate the monetary value of holding $\phi_t$ shares of the underlying asset at time $t$:$$\phi_t S_t = \left( \frac{2e^{-r(T-t)}}{S_t} \right) S_t = 2e^{-r(T-t)}$$Notice how the stochastic asset price $S_t$ cancels out. This means the monetary value of the stock position depends only on time to maturity.
#### Step 2: Valuation of the Money-Market Position 
Next, we calculate the monetary value of holding $\psi_t$ units in the bank account $A_t = e^{rt}$:$$\psi_t A_t = \left[ 2e^{-rT} \left( \int_{0}^{t} \frac{dS_u}{S_u} + (T-t)r - 1 \right) \right] e^{rt}$$By combining the exponential terms $e^{-rT} \cdot e^{rt} = e^{-r(T-t)}$, we get:$$\psi_t A_t = 2e^{-r(T-t)} \left( \int_{0}^{t} \frac{dS_u}{S_u} + (T-t)r - 1 \right)$$We can distribute the $2e^{-r(T-t)}$ term inside the parenthesis:$$\psi_t A_t = 2e^{-r(T-t)} \int_{0}^{t} \frac{dS_u}{S_u} + 2r(T-t)e^{-r(T-t)} - 2e^{-r(T-t)}$$
#### Step 3: Assembly of the Portfolio Value 
Now we substitute the evaluated components back into the total portfolio equation $V_t = L_t + \phi_t S_t + \psi_t A_t$:$$V_t = L_t + \underbrace{2e^{-r(T-t)}}_{\phi_t S_t} + \underbrace{2e^{-r(T-t)} \int_{0}^{t} \frac{dS_u}{S_u} + 2r(T-t)e^{-r(T-t)} - 2e^{-r(T-t)}}_{\psi_t A_t}$$
#### Step 4: Simplification and Conclusion 
We can immediately see that the constant term $+2e^{-r(T-t)}$ from the stock position exactly cancels out the $-2e^{-r(T-t)}$ term from the money-market position:$$V_t = L_t + 2r(T-t)e^{-r(T-t)} + 2e^{-r(T-t)} \int_{0}^{t} \frac{dS_u}{S_u}$$This final expression for $V_t$ is mathematically identical to the decomposition of $\Pi_t$ we proved in Section 1.3.Therefore, we have demonstrated that:$$V_t = \Pi_t$$This confirms that the specified continuous trading strategy involving the log-contract, the underlying asset, and the money market account perfectly replicates the realized variance claim.

This mathematical equality represents the fundamental principle of No-Arbitrage. If $V_t \neq \Pi_t$, an arbitrageur could exploit the difference for a risk-free profit. For instance, if the market price of the variance claim was higher than the cost of our portfolio ($\Pi_t > V_t$), a trader could sell the variance claim to a client for $\Pi_t$ and immediately build the replicating portfolio for $V_t$. They would pocket the difference ($\Pi_t - V_t$) today as a risk-free profit, because the portfolio perfectly guarantees the exact payout needed at maturity $T$. In an efficient market, these opportunities are instantly closed, forcing the theoretical price of the derivative to be exactly equal to the cost of its replicating portfolio.
